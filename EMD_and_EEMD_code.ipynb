```python
from PyEMD import EMD
from PyEMD import Visualisation
import numpy  as np
import pylab as plt

# Define signal
t = np.linspace(0, 1, 200)            #time
s = np.cos(11*2*np.pi*t*t) + 6*t*t    #signal

# Execute EMD on signal
emd = EMD()                           #declaring EMD empirical mode decomposition instance
IMF = emd.emd(s,t)                    #applying emd
IMF,residual = emd.get_imfs_and_residue() #getting IMFs and residue

# Plot results
x = Visualisation(emd_instance=None)  #creating visualization instance
x.plot_imfs(imfs=IMF, residue=residual, t=t, include_residue=True) #plotting IMFs and residue
```

```python
from PyEMD import EEMD
from PyEMD import Visualisation
import numpy as np
import pylab as plt

# Define signal 
t = np.linspace(0, 1, 200)                       #time
S = np.cos(11*2*np.pi*t*t) + 6*t*t               #signal

# Assign EEMD to `eemd` variable
eemd = EEMD()                                    #declaring EEMD ensemble empirical mode decomposition instance

# Execute EEMD on S
eIMFs = eemd.eemd(S, t)                          #applying emd
E_IMF,residual = eemd.get_imfs_and_residue()     #getting IMFs and residue

# Plot results
x = Visualisation(emd_instance=None)             #creating visualization instance
x.plot_imfs(imfs=E_IMF, residue=residual, t=t, include_residue=True)  #plotting IMFs and residue
```
